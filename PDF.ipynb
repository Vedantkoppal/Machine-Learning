```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
num=[]
for i in range(10000):
    num.append(random.randint(1, 5))
num=np.array(num)
```

```python
len(num)
```

```python
num[:5]
```

```python
unique=np.unique(num,return_counts=True)
```

```python
unique[0]
```

```python
plt.bar(unique[0], unique[1], align='center', alpha=0.7,)
plt.xlabel('Unique Values')
plt.ylabel('Counts')
plt.title('Histogram of Unique Values')
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(num, fill=True)
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Kernel Density Estimate (KDE) Plot')
plt.show()
```
